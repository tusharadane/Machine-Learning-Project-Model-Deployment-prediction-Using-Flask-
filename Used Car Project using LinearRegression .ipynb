```python
data.columns
```

```python
x=data.drop(['Car_Name','Selling_Price','Seller_Type','Car_Model'],axis=1)

y=data["Selling_Price"]


x.head()
y.head()
```

```python
print('x_train dataset',len(x_train))
print('y_train dataset',len(y_train))
```

```python
lin_reg.fit(x_train,y_train)

LinearRegression()
```

```python
coef=lin_reg.coef_
coef
```

```python
y_pred=lin_reg.predict(x_test)

y_pred
```

```python
acc=(sum(y_pred)/sum(y_test))
acc
```
